```python
def load_and_preprocess_data(data_folder="processed_data"):
    """Load and preprocess skeleton sequences and labels"""
    sequences = []
    labels = []

    data_folder = Path(data_folder)
    for action, action_id in ACTION_TO_ID.items():
        action_dir = data_folder / action
        if not action_dir.exists():
            continue

        for npy_file in action_dir.glob("*.npy"):
            # Load sequence
            seq = np.load(npy_file)

            # Preprocess: normalize and reshape
            seq = seq.reshape(-1, NUM_JOINTS, 3)
            seq = seq - seq[:, 0:1, :]  # Center to root joint

            # Get temporal boundaries from filename (format: sXX_eXX_start_end.npy)
            parts = npy_file.stem.split('_')
            start_frame = int(parts[-2])
            end_frame = int(parts[-1])

            sequences.append(seq)
            labels.append({
                action: (start_frame, end_frame)
            })

    return sequences, labels
```

```python
def skeleton_to_encoded_image(sequence, output_size=INPUT_SIZE):
    """Convert skeleton sequence to RGB image representation"""
    # Normalize coordinates
    min_val = sequence.min(axis=(0, 1))
    max_val = sequence.max(axis=(0, 1))
    norm_seq = (sequence - min_val) / (max_val - min_val + 1e-8)

    # Create joint heatmaps for each frame
    heatmaps = []
    for frame in norm_seq:
        # Create empty image
        img = np.zeros((output_size[0], output_size[1], 3), dtype=np.float32)

        # Draw joints
        for joint in frame:
            x = int(joint[0] * (output_size[1] - 1))
            y = int(joint[1] * (output_size[0] - 1))
            if 0 <= x < output_size[1] and 0 <= y < output_size[0]:
                # Use z-coordinate for intensity
                intensity = min(max(joint[2], 0.1), 1.0)
                cv2.circle(img, (x, y), 2, (intensity, intensity, intensity), -1)

        heatmaps.append(img)

    # Stack temporal dimension vertically
    final_image = np.vstack(heatmaps)

    # Resize to target size
    final_image = cv2.resize(final_image, output_size)
    return (final_image * 255).astype(np.uint8)
```

```python
def create_fasterrcnn_dataset(sequences, action_labels):
    """Create Detectron2 compatible dataset dictionary"""
    dataset_dicts = []

    for idx, (seq, labels) in enumerate(zip(sequences, action_labels)):
        # Convert to image representation
        img = skeleton_to_encoded_image(seq)
        height, width = img.shape[:2]

        # Prepare record
        record = {
            "file_name": f"sequence_{idx}.jpg",
            "image_id": idx,
            "height": height,
            "width": width,
            "annotations": []
        }

        # Create temporal bounding boxes
        for action, (start, end) in labels.items():
            # Convert frame numbers to image coordinates
            seq_length = seq.shape[0]
            x1 = (start / seq_length) * width
            x2 = (end / seq_length) * width

            record["annotations"].append({
                "bbox": [x1, 0, x2, height],  # (x1, y1, x2, y2)
                "bbox_mode": BoxMode.XYXY_ABS,
                "category_id": ACTION_TO_ID[action],
                "iscrowd": 0
            })

        dataset_dicts.append(record)

    return dataset_dicts
```

```python
def setup_config():
    """Set up Faster R-CNN configuration"""
    cfg = get_cfg()
    cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))

    # Dataset
    cfg.DATASETS.TRAIN = ("skeleton_train",)
    cfg.DATASETS.TEST = ("skeleton_val",)
    cfg.DATALOADER.NUM_WORKERS = 2

    # Model
    cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml")
    cfg.MODEL.ROI_HEADS.NUM_CLASSES = len(ACTION_TO_ID)
    cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5

    # Solver
    cfg.SOLVER.IMS_PER_BATCH = 4
    cfg.SOLVER.BASE_LR = 0.00025
    cfg.SOLVER.MAX_ITER = 1000
    cfg.SOLVER.STEPS = (500, 750)

    # Input
    cfg.INPUT.MIN_SIZE_TRAIN = INPUT_SIZE[0]
    cfg.INPUT.MAX_SIZE_TRAIN = INPUT_SIZE[1]
    cfg.INPUT.MIN_SIZE_TEST = INPUT_SIZE[0]
    cfg.INPUT.MAX_SIZE_TEST = INPUT_SIZE[1]

    return cfg
```

```python
class TemporalFasterRCNN(nn.Module):
    """Custom Faster R-CNN with temporal processing"""
    def __init__(self, cfg):
        super().__init__()
        self.backbone = build_backbone(cfg)
        self.proposal_generator = build_proposal_generator(cfg, self.backbone.output_shape())
        self.roi_heads = build_roi_heads(cfg, self.backbone.output_shape())

        # Temporal processing
        self.temporal_conv = nn.Conv3d(256, 256, kernel_size=(3, 1, 1), padding=(1, 0, 0))

    def forward(self, batched_inputs):
        images = [x["image"].to(self.device) for x in batched_inputs]
        features = self.backbone(images.tensor)

        # Add temporal dimension
        features = {k: v.unsqueeze(-1) for k, v in features.items()}

        # Temporal convolution
        features = {k: self.temporal_conv(v) for k, v in features.items()}

        proposals, _ = self.proposal_generator(images, features, None)
        results, _ = self.roi_heads(images, features, proposals, None)

        return results
```

```python
def register_datasets(dataset_dicts):
    """Register datasets with Detectron2"""
    train_dicts, val_dicts = train_test_split(dataset_dicts, test_size=0.2, random_state=42)

    DatasetCatalog.register("skeleton_train", lambda: train_dicts)
    DatasetCatalog.register("skeleton_val", lambda: val_dicts)
    MetadataCatalog.get("skeleton_train").set(thing_classes=list(ACTION_TO_ID.keys()))

    return train_dicts, val_dicts
```

```python
def train_model(cfg):
    """Train the Faster R-CNN model"""
    os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)

    # Use custom model if needed
    if cfg.MODEL.TEMPORAL:
        model = TemporalFasterRCNN(cfg)
        trainer = DefaultTrainer(cfg, model)
    else:
        trainer = DefaultTrainer(cfg)

    trainer.resume_or_load(resume=False)
    trainer.train()

    return trainer
```

```python
def evaluate_model(trainer, cfg):
    """Evaluate model performance"""
    evaluator = COCOEvaluator("skeleton_val", output_dir=cfg.OUTPUT_DIR)
    val_loader = build_detection_test_loader(cfg, "skeleton_val")
    results = inference_on_dataset(trainer.model, val_loader, evaluator)

    print("Evaluation results:")
    print(json.dumps(results, indent=2))

    return results
```

```python
def main():
    # 1. Load and preprocess data
    print("Loading and preprocessing data...")
    sequences, action_labels = load_and_preprocess_data()

    # 2. Create Detectron2 dataset
    print("Creating dataset...")
    dataset_dicts = create_fasterrcnn_dataset(sequences, action_labels)

    # 3. Register datasets
    print("Registering datasets...")
    train_dicts, val_dicts = register_datasets(dataset_dicts)

    # 4. Set up config
    print("Setting up configuration...")
    cfg = setup_config()

    # 5. Train model
    print("Training model...")
    trainer = train_model(cfg)

    # 6. Evaluate
    print("Evaluating model...")
    results = evaluate_model(trainer, cfg)

    # 7. Save model
    print("Saving model...")
    torch.save(trainer.model.state_dict(), "faster_rcnn_skeleton.pth")

    print("Training complete!")

if __name__ == "__main__":
    main()
```
