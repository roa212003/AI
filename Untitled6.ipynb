```python
import os
import cv2
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
def preprocess_utkinect_sequences(sequences, target_length=30, num_joints=20):
    processed_seqs = []

    for idx, seq in enumerate(sequences):
        if seq.ndim == 2 and seq.shape[1] == num_joints * 3:
            seq = seq.reshape(seq.shape[0], num_joints, 3)
        elif seq.ndim == 3 and seq.shape[1] == num_joints and seq.shape[2] == 3:
            pass
        else:
            raise ValueError(f"Sequence {idx} có shape không đúng: {seq.shape}")

        # 1. Smoothing
        window_size = 3
        smoothed = np.zeros_like(seq)
        for i in range(seq.shape[0]):
            start = max(0, i - window_size // 2)
            end = min(seq.shape[0], i + window_size // 2 + 1)
            smoothed[i] = np.mean(seq[start:end], axis=0)

        # 2. Normalization:
        normalized = smoothed - smoothed[:, 0:1, :]

        # 3. Resize về target_length
        if normalized.shape[0] > target_length:
            # Nếu dài hơn: lấy mẫu đều
            indices = np.linspace(0, normalized.shape[0] - 1, target_length).astype(int)
            resized = normalized[indices]
        else:
            # Nếu ngắn hơn: padding zeros
            pad_len = target_length - normalized.shape[0]
            resized = np.pad(normalized, ((0, pad_len), (0, 0), (0, 0)), mode='constant')

        processed_seqs.append(resized)

    return np.array(processed_seqs)
```

```python
# --- Skeleton to RGB Image Encoding ---
def skeleton_to_encoded_image(sequence, normalize=True):
    num_frames, num_joints, _ = sequence.shape
    encoded = np.zeros((num_frames, num_joints, 3))

    if normalize:
        coord_min = sequence.min(axis=(0, 1))
        coord_max = sequence.max(axis=(0, 1))
        coord_range = np.where(coord_max - coord_min < 1e-8, 1, coord_max - coord_min)

        encoded = 255 * (sequence - coord_min) / coord_range
        encoded = np.floor(encoded).astype(np.uint8)
    else:
        encoded = sequence.copy()

    return encoded
```

```python
# --- CNN Model Definition ---
def build_baseline_cnn(input_shape, num_classes=10):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        MaxPooling2D((2, 2)),

        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),

        Conv2D(128, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),

        Flatten(),
        Dense(256, activation='relu'),
        Dense(128, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
def main():
    import pathlib
    output_folder = pathlib.Path("/content/processed_data")

    action_names = ["walk", "sitDown", "standUp", "pickUp", "carry", "throw", "push", "pull", "waveHands", "clapHands"]
    action_to_idx = {action: idx for idx, action in enumerate(action_names)}

    all_data = []
    all_labels = []

    # Load dữ liệu
    for action in action_names:
        action_dir = output_folder / action
        if not action_dir.exists():
            continue
        for npy_file in action_dir.glob("*.npy"):
            seq = np.load(npy_file)
            all_data.append(seq)
            all_labels.append(action_to_idx[action])

    print("Tổng số sequence ban đầu:", len(all_data))

    # Tiền xử lý
    processed_data = preprocess_utkinect_sequences(all_data, target_length=30)
    all_labels = np.array(all_labels)

    print("Sau tiền xử lý:", len(processed_data), "sequence")

    # Encode và kiểm tra lỗi
    encoded_images = []
    valid_labels = []

    for seq, label in zip(processed_data, all_labels):
        try:
            img = skeleton_to_encoded_image(seq)
            encoded_images.append(img)
            valid_labels.append(label)
        except Exception as e:
            print(f"Bỏ qua sequence do lỗi encode: {e}")
            continue

    encoded_images = np.array(encoded_images)
    valid_labels = np.array(valid_labels)

    print("Encoded images shape:", encoded_images.shape)
    print("Final labels shape:", valid_labels.shape)

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        encoded_images, valid_labels, test_size=0.2, random_state=42, stratify=valid_labels)

    # --- Data augmentation ---
    data_augmentation = tf.keras.Sequential([
        tf.keras.layers.RandomFlip("horizontal"),  # Lật ngẫu nhiên trái/phải
        tf.keras.layers.RandomRotation(0.05),
    ])

    def augment(image):
        image = tf.image.convert_image_dtype(image, tf.float32)  # Đảm bảo chuẩn hóa
        image = data_augmentation(image)
        return image

    AUTOTUNE = tf.data.AUTOTUNE

    # Tạo dataset tf.data từ numpy array
    train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train))
    train_ds = train_ds.shuffle(buffer_size=1000).map(lambda x, y: (augment(x), y), num_parallel_calls=AUTOTUNE).batch(16).prefetch(AUTOTUNE)

    test_ds = tf.data.Dataset.from_tensor_slices((X_test, y_test)).batch(32).prefetch(AUTOTUNE)

    # --- Build và train model ---
    model = build_baseline_cnn(input_shape=(30, 20, 3), num_classes=10)
    model.summary()

    model.fit(train_ds, validation_data=test_ds, epochs=20)

    # --- Đánh giá ---
    test_loss, test_acc = model.evaluate(test_ds, verbose=2)
    print(f"Test Accuracy: {test_acc:.4f}")

    # --- Lưu model ---
    model.save('my_model.h5')

if __name__ == "__main__":
    main()
```
